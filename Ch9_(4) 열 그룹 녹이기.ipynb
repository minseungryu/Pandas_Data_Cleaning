# 열 그룹을 녹이기
- (3)에서 두 그룹의 열을 녹이기 위해서는 melt를 두 번 실행한 뒤, 하나의 데이터프레임으로 병합해야 했음
- wide_to_long 함수를 사용해도 동일한 결과를 얻을 수 있음 👉 기능이 많지만 사용법이 좀 더 복잡
---
- 열 그룹에서 반복되는 부분(weeksworked, colenroct 같은 부분)을 함수에 전달하는 것이 포인트
- 열 이름의 뒷부분(연도)을 값으로 바꿔서 j매개변수로 지정한 열에 채워준다

### 데이터 : NLS 근무 주 수 & 대학 등록 데이터

```python
import pandas as pd

nls97= pd.read_csv('data/nls97f.csv')
nls97.set_index('personid', inplace=True)
```

```python
# 근무 주 수와 대학등록 데이터 일부 확인
weeksworkedcols = ['weeksworked00', 'weeksworked01', 'weeksworked02', 'weeksworked03', 'weeksworked04']
colenrcols = ['colenroct00', 'colenroct01', 'colenroct02', 'colenroct03', 'colenroct04']
```

```python
nls97.loc[nls97.originalid.isin([1, 2]), ['originalid'] + weeksworkedcols + colenrcols].T
```

### 📍wide_to_long 함수 실행
- stubnames에 리스트를 전달 ->> 변환하려는 열 그룹 지시(리스트의 각 아이템과 같은 이름으로 시작하는 모든 열 선택)
- i 매개변수로 id 변수 전달
- j 매개변수로 열 이름 뒤에 붙어있던 "00, 01등에 기반한" 열 이름(year) 지정

```python
workschool = pd.wide_to_long(nls97[['originalid']+weeksworkedcols+colenrcols],
               stubnames=['weeksworked', 'colenroct'],
               i=['originalid'], j='year').reset_index()
```

```python
workschool['year'] = workschool.year + 2000
```

```python
workschool = workschool.sort_values(['originalid', 'year'])
```

```python
workschool.set_index(['originalid'], inplace = True)
```

```python
workschool.head(10)
```

### 추가 학습
- 위의 예시에서는 stubnames 열의 뒷부분이 00~04로 동일하지만 항상 이렇게 되어 있으리라는 법은 없다.
- 한쪽 그룹에는 뒷부분에 글자가 붙어있고, 다른 그룹에는 그렇지 않다면, 후자는 누락값으로 처리된다.
- 이러한 예를 weeksworked03을 빼고 weeksworked05를 추가해서 확인해본다.

```python
weeksworkedcols = ['weeksworked00', 'weeksworked01', 'weeksworked02', 'weeksworked04', 'weeksworked05']
```

```python
workschool = pd.wide_to_long(nls97[['originalid']+weeksworkedcols+colenrcols],
                            stubnames=['weeksworked', 'colenroct'],
                            i=['originalid'], j='year').reset_index()
```

```python
workschool['year'] = workschool['year']+2000
```

```python
workschool = workschool.sort_values(['originalid', 'year'])
workschool.set_index(['originalid'], inplace = True)
workschool.head(12)
```

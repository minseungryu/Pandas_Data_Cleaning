# groupby를 사용해 데이터프레임의 분석 단위를 바꾸기

```python
import pandas as pd
coviddaily = pd.read_csv('data/coviddaily720.csv', parse_dates=['casedate'])
ltbrazil = pd.read_csv('data/ltbrazil.csv')
```

### 코로나19 데이터를 국가별 일일 데이터 ->> 전체 국가의 일일 데이터로 변환

```python
coviddailytotals = coviddaily.loc[coviddaily.casedate.between('2020-02-01', '2020-07-12')].groupby(['casedate'], as_index=False)[['new_cases', 'new_deaths']].sum()
```

```python
coviddailytotals.head(10)
```

### 브라질의 기상 관측소별 평균 온도 데이터프레임 생성
- 온돗값이 누락된 행을 제거한 뒤 데이터 표시

```python
ltbrazil = ltbrazil.dropna(subset=['temperature'])
```

```python
ltbrazil[['station', 'year', 'month', 'temperature', 'elevation', 'latabs']]
```

```python
ltbrazilavgs = ltbrazil.groupby(['station'], as_index=False).agg({'latabs':'first', 'elevation':'first', 'temperature':'mean'})
```
